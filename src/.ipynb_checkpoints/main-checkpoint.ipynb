```python
import pandas as pd
import numpy as np
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from pprint import pprint
from sklearn.model_selection import cross_val_score
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import plot_confusion_matrix
```

```python
data = pd.read_csv('../data/cases_train_processed.csv')
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
data['sex'] = labelencoder.fit_transform(data['sex'])
data['province'] = labelencoder.fit_transform(data['province'])
data['country'] = labelencoder.fit_transform(data['country'])
data['date_confirmation'] = labelencoder.fit_transform(data['date_confirmation'])
```

### Split dataset

```python
training_data, validation_data = np.split(data.sample(frac=1),
                                              [int(0.8 * len(data))])
x_columns = [x for x in training_data.columns if x!='outcome']
X_train = training_data[x_columns]
X_validation = validation_data[x_columns]
y_train = training_data['outcome']
y_validation = validation_data['outcome']
```

### Confusion Matrix Ploter

```python
def plot_confusion_matrix(y_true, y_pred):
    matrix = confusion_matrix(y_true, y_pred)
    matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
    # Build the plot
    plt.figure(figsize=(16,7))
    sns.set(font_scale=1.4)
    sns.heatmap(matrix, annot=True, annot_kws={'size':10},
                cmap=plt.cm.Greens, linewidths=0.2)

    # Add labels to the plot
    class_names = ['deceased','hospitalized','nonhospitalized', 'recovered']
    tick_marks = np.arange(len(class_names)) + 0.5
    plt.xticks(tick_marks, class_names, rotation=0)
    plt.yticks(tick_marks, class_names, rotation=0)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix for Random Forest Model')
    plt.show()
```

##### Accuracy Score
   - Accuray score measures how many labels the model got right out of the total number of predictions. 
   - This model has an accuracy of 88.66% on the training dataset, it may not look too bad at first glance. 
   - However, accuracy is not a great measure of classifier performance when the classes are imbalanced. This means there can be scenario that labels with small number of occurence (such as deceased in this case) are not correctly predicted.
   - More information are needed for us to well understand the real performance of this model. Did it perform equally well for each class? Were there any pairs of classes it found especially hard to distinguish?

```python
accuracy_score(y_train, pred_train)
```

```python
labels = ['deceased', 'recovered', 'hospitalized', 'nonhospitalized']
counter = [
    y_train.tolist().count('deceased'),
    y_train.tolist().count('recovered'),
    y_train.tolist().count('hospitalized'),
    y_train.tolist().count('nonhospitalized')]
plt.figure(figsize=(8,4))
plt.title('Barplot of Labels')
plt.bar(labels, counter)
```

##### Confusion Matrix
- A confusion matrix is a way to express how many of a classifier’s predictions were correct, and when incorrect, where the classifier got confused.
- It’s easy to see that our classifier struggled at classifying <i>deceased</i> label. About half the time, deceased were mislabeled as <i> hospitalized</i>.
- Also, about a quarter of time, <i>recoverd</i> is incorrectly predicted as <i>hospitalized</i>.

```python
plot_confusion_matrix(y_train, pred_train)
```

##### Classification report (Precision, Recall, F1 Score)
- All labels has a high precision, meaning that the model was careful to avoid classify things that are not label X into label X.
- <i> decreased </i> has a very low recall, meaning a lot of <i> decreased </i> were not correctly classified.
- f1-score reflect the imbalance

```python
print(classification_report(y_train, pred_train))
```

##### Summary
On its own, a classification report tells us generally what kind of errors the model made, but it doesn’t give us specifics. The confusion matrix tells us exactly where mistakes were made, but it doesn’t give us summary metrics like precision, recall, or F1 score. Using both of these can give us a much more nuanced understanding of how our model performs, going far beyond what an accuracy score can tell us and avoiding some of its pitfalls.

## Task3 - Overfitting
- From the evalution above, we notice that the model works better with training dataset than validation dataset. Therefore, the overfitting exist.
- We would like to tune hyperparameters through cross validation.
- Reference: https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74

### Get parameters

```python
pprint(rfModel.get_params())
```

### n_estimators
in general the more trees the less likely the algorithm is to overfit. So try increasing this. The lower this number, the closer the model is to a decision tree, with a restricted feature set.

```python
ns = [100, 200, 400, 800]
for n in ns:
    rfTestModel = RandomForestClassifier(n_estimators=n)
    print(f'n_estimators={n}')
    scores = cross_val_score(rfTestModel, X_validation, y_validation, cv=5)
    print(scores)
    print()
```

### max_features
This determines how many features each tree is randomly assigned. The smaller, the less likely to overfit, but too small will start to introduce under fitting.

```python
ns = [4, 8, 10, 12]
for n in ns:
    rfTestModel = RandomForestClassifier(max_features=n)
    print(f'max_features={n}')
    scores = cross_val_score(rfTestModel, X_validation, y_validation, cv=5)
    print(scores)
    print()
```

### min_samples_leaf
- This determines the minimum number of sample a leaf node should have before further split. There can be overfitting if min_samples_leaf is too small.
- From the experiment we found increases min_samples_leaf alone does not help solve overfitting

```python
rfModel.min_samples_leaf
```

```python
min_leafs = [1, 2, 4, 8, 16]
for min_leaf in min_leafs:
    rfTestModel = RandomForestClassifier(min_samples_leaf=min_leaf)
    print(f'min_samples_leaf={min_leaf}')
    scores = cross_val_score(rfTestModel, X_validation, y_validation, cv=5)
    print(scores)
    print()
```

```python
num_leaves = 200
n_estimators = 200
gbdt_lgb = LGBMClassifier(boosting_type = 'gbdt', num_leaves = num_leaves, n_estimators=n_estimators).fit(X_train, y_train)
dart_lgb = LGBMClassifier(boosting_type = 'dart', num_leaves = num_leaves,n_estimators=n_estimators).fit(X_train, y_train)
goss_lgb = LGBMClassifier(boosting_type = 'goss', num_leaves = num_leaves,n_estimators=n_estimators).fit(X_train, y_train)
```

### Task3 - Overfitting

```python
acc_score =  pd.DataFrame(columns = ['Training Set', 'Test Set'])
for i in [1,10,15,25,30]:
     lgbm = LGBMClassifier(num_leaves = 1000, min_child_samples = i, n_estimators = 1
                         ).fit(X_train, y_train)
     acc_score.loc[i] = [accuracy_score(y_train, lgbm.predict(X_train)), accuracy_score(y_validation, lgbm.predict(X_validation))]
```

```python
plt.title('LGB Overfitting')
plt.plot(acc_score['Training Set'], label = 'Training Set')
plt.plot(acc_score['Test Set'], label = 'Test Set')
plt.legend()
plt.xlabel('Min Child Samples')
plt.ylabel('Classification  accuracy')
plt.show()
plt.savefig('../plots/lgbm_overfitting.png')
```
